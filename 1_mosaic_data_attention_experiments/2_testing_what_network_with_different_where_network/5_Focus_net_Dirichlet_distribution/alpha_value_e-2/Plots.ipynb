```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
train_loss_all = np.load("train_loss.npy")
```

```python
%matplotlib inline

fig = plt.figure()
for i,j in enumerate(train_loss_all):
   plt.plot(j,label ="dataset "+str(i+1))
    

plt.xlabel("Epochs")
plt.ylabel("Training_loss")
plt.title("alpha_value_0.02")
plt.legend()
fig.savefig("Figure.pdf")
fig.savefig("Figure.png")
```
