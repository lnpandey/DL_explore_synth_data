```python
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
mu1 , sigma1 = 0 , 1
mu2 , sigma2 = 0 , 1
radius1=2
delta = 0.4
radius2 = radius1+delta
radius3 = radius1-delta
num_of_feature = 2 
num_of_data_points = 100

def generate_dataPoints_on_sphere(no_features,no_data,mu,sigma,radius,plus_or_minus):
    data=[]
    for i in range(no_data):
        s = np.random.normal(mu, sigma, no_features)
        s_dash = copy.deepcopy(s)
        s=s**2
        sum_all=np.sum(s)
        s= radius*s_dash/np.sqrt(sum_all)
        s=list(s)
        s.insert(no_features,plus_or_minus)
        data.append(s)
    return np.array(data)

positive_data = generate_dataPoints_on_sphere(num_of_feature,num_of_data_points,mu1,sigma1,radius1,1)
negative_data1 = generate_dataPoints_on_sphere(num_of_feature,num_of_data_points,mu2,sigma2,radius2,-1)
negative_data2 = generate_dataPoints_on_sphere(num_of_feature,num_of_data_points,mu2,sigma2,radius3,-1)
# print(positive_data)
# print(negative_data)
```

```python
x1 = positive_data[:,0]
y1 = positive_data[:,1]
# print(x1,y1)

x2 = negative_data1[:,0]
y2 = negative_data1[:,1]
# print(x2,y2)

x3 = negative_data2[:,0]
y3 = negative_data2[:,1]
# print(x3,y3)
```

```python
plt.scatter(x1, y1, alpha=0.6)
plt.scatter(x2, y2 , alpha=0.6)
plt.scatter(x3, y3 , alpha=0.6)
plt.title('Scatter plot pythonspot.com')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
